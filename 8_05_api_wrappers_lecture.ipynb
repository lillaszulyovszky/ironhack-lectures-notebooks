````python
artists = ['Marc Rebillet',
           'Modeselektor',
           'Who Made Who',
           'Hildegard von Blingin']
````

````python
# phine with append
my_20_artists = []
for artist in artists:
    my_20_artists.append(sp.search(q={artist}, limit=50)) 

# marian list comprehension
my_20_artists = [sp.search(q=artist, limit=50) for artist in artists]
````

### second part of the activity

````python
artists = ['Marc Rebillet',
           'Modeselektor',
           'Who Made Who',
           'Hildegard von Blingin']

def artist_to_dict(artists):
    return {artist:sp.search(q=artist, limit=50) for artist in artists}

len(artist_to_dict(artists))
````

````python
# dictionary comprehension
{str(key):key**2 for key in range(10)}#

# same thing with a for loop
dct = {}
for key in range(10):
    dct[str(key)] = key**2

dct
````
