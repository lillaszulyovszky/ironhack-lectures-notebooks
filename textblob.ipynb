```python
import pandas as pd
from textblob import TextBlob
```

```python
# read excel file and only the Open Comments tab
xls = pd.ExcelFile('Comments on the survey-2.xlsx')
df1 = pd.read_excel(xls, 'Open Comments')
```

```python
# only include these 3 columns
df1 = df1[['Open Comments', 'Decisions', 'Actions']]
```

```python
df1.head(30)
```

```python
def sentiment(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None
```

```python
df1['poparity']=df1['Open Comments'].apply(sentiment).apply(lambda x:x[0])
```

```python
df1['subjectivity']=df1['Open Comments'].apply(sentiment).apply(lambda x:x[1])
```
